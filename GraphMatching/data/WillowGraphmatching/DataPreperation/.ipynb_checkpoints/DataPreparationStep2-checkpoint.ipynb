```python
import pickle 
import numpy as np
import itertools

#from DataPreparation.ipynb
toSave = pickle.load( open( "SparsetestData4.p", "rb" ) )

```

```python



entries=len(toSave)//3
print(entries)
subset=np.arange(4)# 4 keypoints were chosen
```

```python
import itertools

#Systematicaly optimization for parameters that give the best brute force solution


for t in   range(101):
        correct=0

        eta=t/100 #2 
        tau=0#82/100
        for s in range(entries-1):

            Input= toSave[-2-3*s]
            sim= Input[0]
            dirMatrix=Input[1][0]
            disMatrix=Input[1][1]
  

            simMatrix= np.diag(np.ravel(sim))



            x =- (1-eta)*simMatrix*tau- (1-tau)*(eta*(np.exp(-dirMatrix**2/ np.var(dirMatrix))-1)+(1-eta)*(np.exp(-disMatrix**2/ np.var(disMatrix))-1))




            def toPerVec(P):
                n=len(P)
                Mat = np.zeros((n,n))
                for k in range(n):
                    Mat[P[k],k]=1

                return np.ravel(Mat, order='F')

            dim=len(subset)
            energy= +1.e+5
            permutation=np.arange(dim)
            for perm in   list(itertools.permutations(np.arange(dim))):

                pvec= toPerVec(perm)
                energynew= pvec.T@x @pvec
                if energynew<energy:
                    permutation= perm
                    energy= energynew
      
            perm=toSave[-1-3*s]
            pvec= toPerVec(perm)
            energynew= pvec.T@x @pvec
            if np.array_equal(perm,permutation):
                correct+=1

        print(correct/entries)
        print(t)
```

```python
def PermutationToBinary(Perm):
    m= int( np.ceil(np.log2(Perm.shape[0])))
    result=     (((Perm[:,None] & (1 << np.arange(m)))) > 0).astype(int)
    result=result.reshape((-1))
    return result




def toPerVec(P):
        #P is a permutation table
        #returns a vectorized permutation 
                n=len(P)
                Mat = np.zeros((n,n))
                for k in range(n):
                    Mat[P[k],k]=1

                return np.ravel(Mat, order='F')



def genData2():
        #Write data in a dictionary with Input and Output keys

        correct=0
        eta=2/100
        tau=82/100#96/100
        dic = {'Input' : [], 'Output' : []}
        number= 40*141
        for s in range(number):
            Input= toSave[-2-3*s]
            sim= Input[0]
            dirMatrix=Input[1][0]
            disMatrix=Input[1][1]


            simMatrix= np.diag(np.ravel(sim))



            x =- (1-eta)*simMatrix*tau- (1-tau)*(eta*(np.exp(-dirMatrix**2/ np.var(dirMatrix))-1)+(1-eta)*np.exp(-disMatrix**2/ np.var(disMatrix))-1)





            
            dim=len(subset)
            energy= +1.e+5
            permutation=np.arange(dim)
            for perm in   list(itertools.permutations(np.arange(dim))):

                pvec= toPerVec(perm)
                energynew= pvec.T@x @pvec
                if energynew<energy:
                    permutation= perm
                    energy= energynew

            permgt=toSave[-1-3*s]
            pvec= toPerVec(permgt)
            energynew= pvec.T@x @pvec
        

            if np.array_equal(permgt,permutation):
                correct+=1

            dic['Input'].append(np.ravel(x))
            dic['Output'].append(PermutationToBinary(np.array(permgt)))
        print(correct/number)
            
        return dic
```
