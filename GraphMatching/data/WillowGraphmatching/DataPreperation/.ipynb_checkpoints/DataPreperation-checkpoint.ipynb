```python
from sklearn.neighbors import NearestNeighbors
import numpy as np


def computeQuadratic(pointsA, pointsB):
    n= len(pointsA)
    quadrCosts1= np.zeros((n**2, n**2))
    quadrCosts2= np.zeros((n**2, n**2))
    
    
    
    nbrs = NearestNeighbors(n_neighbors=n, algorithm='ball_tree').fit(pointsA)
    distances, indices1 = nbrs.kneighbors(pointsA)
    nbrs = NearestNeighbors(n_neighbors=n, algorithm='ball_tree').fit(pointsB)
    distances, indices2 = nbrs.kneighbors(pointsB)
    
    
    
    dis1= np.zeros((n,n))
    dis2= np.zeros((n,n))
    diff1= np.zeros((n,n,2))
    diff2= np.zeros((n,n,2))
    

    
   
            
            
            
    for k in range(n):
        for l in indices1[k]:
            
            dis1[k,l]= np.linalg.norm(pointsA[k]- pointsA[l])
            diff1[k,l]= pointsA[k]- pointsA[l]
            
    for k in range(n):
        for l in indices2[k]:       
            
            dis2[k,l]= np.linalg.norm(pointsB[k]- pointsB[l])
            diff2[k,l]= pointsB[k]- pointsB[l]
            
    for k in range(n):
        for l in indices1[k]:
            for Pk in range(n):
                for Pl in indices2[Pk]:
                    if k!=l and Pk != Pl:

                        quadrCosts1[k*n + Pk , l*n + Pl ]= (dis1[k,l]- dis2[Pk,Pl])/ (dis1[k,l]+ dis2[Pk,Pl])
                        quadrCosts2[k*n + Pk , l*n + Pl ]= np.arccos(diff1[k,l]/ dis1[k,l] @diff2[Pk,Pl]/ dis2[Pk,Pl])

    return quadrCosts1, quadrCosts2 
```
